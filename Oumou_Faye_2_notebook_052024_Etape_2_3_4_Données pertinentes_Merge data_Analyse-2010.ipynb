```python
# Importer les librairies nécéssaires pour sélectionner les données pertinentes et éffectuer le merge des fichiers suivants:

#'data_EdStatsCountry' Statistiques par Pays
#'data_EdStatsData' Indicateurs par Pays et par année

# Importation de la librairie python **Pandas** utilisée pour la manipulation et l'analyse de données
import pandas as pd

# Importation de la librairie python **Numpy** utilisée pour les calculs scientifique et la manipulation des tableaux
import numpy as np
```

```python
# Importation et lecture du fichier 'EdStatsCountry.csv' et du fichier 'EdStatsData.csv'
# (Utilisation de la librairie Pandas)

data_EdStatsCountry = pd.read_csv('EdStatsCountry.csv')

data_EdStatsData = pd.read_csv('EdStatsData.csv')
```

```python
# Afficher la liste des colonnes du fichier 'EdStatsCountry.csv' pour selectionner les données pertinentes

data_EdStatsCountry.columns
```

```python
# Conserver uniquement les colonnes pertinentes avant d'effectuer le merge avec le fichier 'EdStatsData' pour effectuer l'analyse à postériorie 
df1_colonnes_a_conserver = ['Country Code', 'Region', 'Income Group']

# Filtrage du dataframe pour ne conserver que les colonnes spécifiées
data_EdStatsCountry = data_EdStatsCountry[df1_colonnes_a_conserver]

# Afficher le dataframe1 à jour et vérrifier la présence des colonnes à conserver
data_EdStatsCountry.head(5)
```

```python
# Afficher la liste des colonnes du fichier 'data_EdStatsData' pour selectionner les données pertinentes

data_EdStatsData.columns
```

```python
# Calculer le pourcentage de valeurs manquantes pour chaque colonne
missing_percentage = data_EdStatsData.isnull().mean() * 100

# Trier les résultats par ordre croissant
missing_percentage_sorted = missing_percentage.sort_values(ascending=True)

# Afficher le résultat
print("Pourcentage de valeurs manquantes par colonne (ordonné par ordre croissant) :")
print(missing_percentage_sorted)

#Optimiser l'affichage de l'output des codes
pd.set_option('display.max_columns', 70)
pd.set_option('display.max_rows', 200)
pd.set_option('display.precision', 5)
pd.set_option('display.width', 1000)
pd.set_option('display.max_colwidth',1000)
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
# Conserver uniquement les colonnes du fichier 'data_EdStatsData' pertinentes avant d'effectuer le merge avec le fichier 'data_EdStatsCountry' pour effectuer l'analyse univariée et bivariée à postériorie 
df2_colonnes_a_conserver = ['Country Code', 'Country Name','Indicator Code', '2010']

# Filtrage du dataframe pour ne conserver que les colonnes spécifiées (analyse sur l'année 2010 car le taux de remplissage des valeurs est le plus élevé)
data_EdStatsData = data_EdStatsData[df2_colonnes_a_conserver]

# Afficher le fichier data_EdStatsData à jour et vérrifier la présence des colonnes à conserver
data_EdStatsData
```

<center><h1>Etape #3 Création du fichier à analyser (merge du fichier 'data_EdStatsCountry' avec le fichier 'data_EdStatsData' (Valeur année 2010) </h1></center>

```python
# Creer le merge du dataframe 'data_EdStatsCountry' avec le dataframe 'data_EdStatsData' avec la colonne commune 'Country Code'

data_to_analyse = pd.merge(data_EdStatsCountry, data_EdStatsData, on='Country Code')

# Display the merged DataFrame
data_to_analyse
```

```python
# Utilisation de la méthode pivot_table pour transformer le DataFrame et afficher les indicateurs en colonnes
data_to_analysevf = data_to_analyse.pivot_table(
    index=['Country Code','Country Name', 'Region', 'Income Group'], 
    columns='Indicator Code', 
    values='2010').reset_index()

# Affichage du DataFrame pivoté
data_to_analysevf
```

```python
# Afficher les statistiques descriptives des groupes de revenus par Pays
describe_income_group = finaldftoanalyse['Income Group'].value_counts()

# Afficher les résultats
print(describe_income_group)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Compter le nombre de pays dans chaque groupe de revenu
income_group_counts = finaldftoanalyse['Income Group'].value_counts()

# Calculer le nombre de pays pour chaque groupe de revenu
grouped_df = finaldftoanalyse.groupby('Income Group')['Country Name'].apply(list)

# Préparer les données pour le graphique
data_for_plot = pd.DataFrame({
    'Income Group': income_group_counts.index,
    'Number of Countries': income_group_counts.values,
    'Countries': [', '.join(countries) for countries in grouped_df]
})

# Créer le graphique à barres horizontales
fig, ax = plt.subplots(figsize=(12, 8))

# Tracer le graphique à barres horizontales
bars = ax.barh(data_for_plot['Income Group'], data_for_plot['Number of Countries'], color='green')

# Ajouter les annotations de valeur sur les barres
for bar, value in zip(bars, data_for_plot['Number of Countries']):
    ax.text(bar.get_width() + bar.get_width()*0.02, bar.get_y() + bar.get_height()/2, 
            f'{value}', va='center', fontsize=12, fontweight='bold')

# Ajouter des labels et un titre
ax.set_title('Groupes de revenus par nombre de pays (2010)', fontsize=14)
ax.set_xlabel('Nombre de pays', fontsize=12)
ax.set_ylabel('Groupe de revenu', fontsize=12)
ax.tick_params(axis='x', labelsize=12)  # Taille de police ajustée pour l'axe des x
ax.tick_params(axis='y', labelsize=12)  # Taille de police ajustée pour l'axe des y

# Créer un sous-graphique pour la légende avec une boîte de défilement simulée
fig.subplots_adjust(right=0.75)  # Ajuster l'espace pour la légende
legend_ax = fig.add_axes([0.76, 0.1, 0.2, 0.8])  # Position et taille du subplot pour la légende
legend_ax.axis('off')

# Ajouter le texte de la légende en plusieurs parties si nécessaire
y_offset = 1
max_lines = 20  # Nombre maximal de lignes visibles
for i, (group, countries) in enumerate(zip(data_for_plot['Income Group'], data_for_plot['Countries'])):
    if i >= max_lines:
        break
    legend_ax.text(0, y_offset - (i * 0.02), f'{group}: {countries}', fontsize=10, va='top', ha='left')

# Ajouter une barre de défilement ou une mention "Voir plus" si nécessaire
if len(data_for_plot) > max_lines:
    legend_ax.text(0, y_offset - (max_lines * 0.02) - 0.05, '... Voir plus', fontsize=10, va='top', ha='left', color='blue')

plt.show()
```

```python
# Matrice de corrélation 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Liste des colonnes d'intérêt
colonnes = ['Country Code', 'Country Name', 'Region', 'Income Group', 
            'NY.GDP.PCAP.PP.KD', 'SE.XPD.TOTL.GD.ZS', 'IT.NET.USER.P2', 
            'SE.TER.ENRL', 'UIS.E.3', 'SE.TER.TCHR', 'UIS.OE.56.40510']

# Filtrer les colonnes numériques
numerical_columns = [col for col in colonnes if col in finaldftoanalyse.columns and finaldftoanalyse[col].dtype in ['float64', 'int64']]

# Calculer la matrice de corrélation
correlation_matrix = finaldftoanalyse[numerical_columns].corr()

# Visualiser la matrice de corrélation
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, vmin=-1, vmax=1)
plt.title('Matrice de Corrélation des Indicateurs (2010)')
plt.show()
```
